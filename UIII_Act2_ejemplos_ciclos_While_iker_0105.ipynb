#### ejemplos de ciclo while


```python
limite = int(input("¿Hasta qué número quieres contar? ")) # 1. Captura el límite y lo convierte a entero.
i = 1                                                   # 2. Inicializa la variable de control en 1.
while i <= limite:                                      # 3. Evalúa si el contador es menor o igual al límite.
    print(i)                                            # 4. Imprime el valor actual de la iteración.
    i += 1                                              # 5. Incrementa el contador para evitar un bucle infinito.
```

```python
nota = -1                                           # 1. Inicializa con un valor fuera de rango para entrar al ciclo.
while nota < 0 or nota > 10:                        # 2. Condición lógica: repite si la nota es inválida.
    nota = float(input("Ingresa una nota (0-10): ")) # 3. Solicita y convierte la entrada a decimal.
    if nota < 0 or nota > 10:                       # 4. Verifica si el dato sigue siendo erróneo.
        print("Error: La nota debe estar entre 0 y 10.") # 5. Informa al usuario sobre el error.
print(f"Nota registrada con éxito: {nota}")         # 6. Mensaje final al obtener un dato correcto.
```

```python
secreto = 7                                         # 1. Define el número que el usuario debe encontrar.
intento = 0                                         # 2. Inicializa la variable para capturar la entrada.
while intento != secreto:                           # 3. Repite mientras el intento sea diferente al secreto.
    intento = int(input("Adivina el número (1-10): ")) # 4. Solicita el número al usuario.
    if intento != secreto:                          # 5. Compara el valor para dar retroalimentación.
        print("Incorrecto, intenta de nuevo.")      # 6. Indica que falló para motivar la repetición.
print("¡Felicidades! Has acertado.")                # 7. Mensaje de éxito al salir del bucle.8
```

```python
continuar = "s"                                     # 1. Inicializa la variable de control con 's' (sí).
while continuar == "s" or continuar == "S":         # 2. Evalúa si el usuario desea seguir operando.
    precio = float(input("Precio del producto: "))  # 3. Captura el valor base.
    iva = precio * 0.16                             # 4. Realiza el cálculo aritmético del impuesto.
    print(f"El IVA de este producto es: {iva}")      # 5. Muestra el resultado al usuario.
    continuar = input("¿Calcular otro? (s/n): ")    # 6. Solicita entrada para decidir si repite el ciclo.
```
