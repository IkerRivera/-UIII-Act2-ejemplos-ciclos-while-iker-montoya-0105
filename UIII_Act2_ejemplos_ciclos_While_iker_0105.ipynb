#### ejemplos de ciclo while


```python
# Ejemplo 1: Suma interactiva
num1 = float(input("Ingresa el primer número: ")) # 1. Solicita el primer dato y lo convierte a decimal.
num2 = float(input("Ingresa el segundo número: ")) # 2. Solicita el segundo dato y lo convierte a decimal.
resultado = num1 + num2 # 3. Realiza la operación aritmética de suma.
print(f"La suma es: {resultado}") # 4. Muestra el resultado final formateado al usuario
```

```python
# Ejemplo 2: Condicional simple
edad = int(input("¿Cuántos años tienes? ")) # 1. Captura la edad y la convierte a número entero.
if edad >= 18: # 2. Evalúa si el valor es mayor o igual a 18 (operador relacional).
    print("Eres mayor de edad.") # 3. Se ejecuta si la condición anterior es verdadera.
else: # 4. Define el camino alternativo si la condición es falsa.
    print("Eres menor de edad.") # 5. Mensaje para cuando el usuario tiene menos de 18 años.
```

```python
# Ejemplo 3: Cálculo geométrico interactivo
largo = float(input("Entre el largo del rectángulo: ")) # 1. Pide la base y la convierte a decimal.
ancho = float(input("Entre el ancho del rectángulo: ")) # 2. Pide la altura y la convierte a decimal.
area = largo * ancho # 3. Multiplica ambas variables para obtener la superficie.
print(f"El área del rectángulo es: {area}") # 4. Informa al usuario el valor calculado del área.
```

```python
# Ejemplo 4: Ciclo controlado por el usuario
suma = 0 # 1. Inicializa la variable acumuladora en cero.
while True: # 2. Inicia un bucle infinito que se controla internamente.
    dato = int(input("Ingresa un número (negativo para salir): ")) # 3. Captura el número.
    if dato < 0: # 4. Verifica si el número es el centinela (negativo).
        break # 5. Rompe el ciclo y sale inmediatamente si se cumple el 'if'.
    suma += dato # 6. Acumula el valor de 'dato' en la variable 'suma'.
print(f"La suma total es: {suma}") # 7. Muestra el total acumulado después de salir del ciclo.
```

```python
# Ejemplo 5: Generador de tablas con rango
base = int(input("¿De qué número quieres ver la tabla? ")) # 1. Captura el número base deseado.
for i in range(1, 11): # 2. Itera desde el 1 hasta el 10 usando la función range.
    resultado = base * i # 3. Calcula el producto del número base por el iterador actual.
    print(f"{base} x {i} = {resultado}") # 4. Muestra cada línea de la tabla de forma ordenada.
```
